```python
import pandas as pd
import numpy as np
from utils import generate_random_sales
import os
os.makedirs("data", exist_ok=True)
```

```python
dates = pd.date_range(start="2025-01-01", periods=12, freq="MS")
dates
```

```python
product_a = generate_random_sales(50, 100, 12, seed=1)
product_b = generate_random_sales(30, 80, 12, seed=2)
product_c = generate_random_sales(20, 60, 12, seed=3)
product_d = generate_random_sales(10, 50, 12, seed=4)
```

```python
initial_df = pd.DataFrame({
    "Date": dates,
    "Product_A": product_a,
    "Product_B": product_b,
    "Product_C": product_c,
    "Product_D": product_d
})

initial_df.to_csv("data/initial.csv", index=False)
initial_df.head()
```

```python
df = initial_df.copy()
df["Month"] = df["Date"].dt.strftime("%Y-%m")
df = df[["Month", "Product_A", "Product_B", "Product_C", "Product_D"]]
```

```python
df["Total_Sales"] = df[["Product_A","Product_B","Product_C","Product_D"]].sum(axis=1)
df["Average_Sales"] = df[["Product_A","Product_B","Product_C","Product_D"]].mean(axis=1)
df["Month_over_Month_Growth"] = df["Total_Sales"].pct_change().fillna(0)*100
df["Month_over_Month_Growth"] = df["Month_over_Month_Growth"].round(2)
def get_quarter(month):
    m = int(month.split("-")[1])
    if 1 <= m <= 3: return "Q1"
    if 4 <= m <= 6: return "Q2"
    if 7 <= m <= 9: return "Q3"
    return "Q4"

df["Quarter"] = df["Month"].apply(get_quarter)
```

```python
def get_quarter(month):
    m = int(month.split("-")[1])
    if 1 <= m <= 3: return "Q1"
    if 4 <= m <= 6: return "Q2"
    if 7 <= m <= 9: return "Q3"
    return "Q4"

df["Quarter"] = df["Month"].apply(get_quarter)
```

```python
prod_cols = ["Product_A","Product_B","Product_C","Product_D"]

df["Max_Sales_Product"] = df[prod_cols].idxmax(axis=1)
df["Min_Sales_Product"] = df[prod_cols].idxmin(axis=1)
```

```python
df.to_csv("data/final.csv", index=False)
df.head(12)
```

```python
pivot_avg = df.groupby("Quarter")[prod_cols + ["Total_Sales"]].mean()
pivot_sum = df.groupby("Quarter")[prod_cols + ["Total_Sales"]].sum()

df.to_csv("data/output.csv", index=False)
pivot_avg, pivot_sum
```

```python
best_month = df.loc[df["Total_Sales"].idxmax(), "Month"]
best_product = df[prod_cols].sum().idxmax()
best_quarter = df.groupby("Quarter")["Total_Sales"].sum().idxmax()

best_month, best_product, best_quarter
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

months_labels = df["Month"].tolist()

plt.figure(figsize=(10,5))

for col in ["Product_A", "Product_B", "Product_C", "Product_D"]:
    plt.plot(months_labels, df[col], marker='o', label=col)

plt.xticks(rotation=45)
plt.title("Ventes mensuelles par produit")
plt.xlabel("Mois")
plt.ylabel("Units vendues")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

bottom = np.zeros(len(df))

for col in ["Product_A", "Product_B", "Product_C", "Product_D"]:
    plt.bar(months_labels, df[col], bottom=bottom, label=col)
    bottom += df[col].values

best_idx = df["Total_Sales"].idxmax()

plt.annotate("Best Month",
             xy=(best_idx, df.loc[best_idx,"Total_Sales"]),
             xytext=(best_idx, df["Total_Sales"].max()*1.05),
             arrowprops=dict(arrowstyle="->", lw=1.5))

plt.xticks(rotation=45)
plt.title("Ventes mensuelles totales (Barres empiles)")
plt.ylabel("Units vendues")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
heat_df = df.set_index("Month")[["Product_A", "Product_B", "Product_C", "Product_D"]]

sns.heatmap(heat_df.T, annot=True, fmt="d", cbar_kws={'label': 'Units Sold'})
plt.title("Heatmap: Ventes mensuelles par produit")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

melted = df.melt(id_vars=["Month"], 
                 value_vars=["Product_A","Product_B","Product_C","Product_D"], 
                 var_name="Product", 
                 value_name="Units")

sns.boxplot(x="Product", y="Units", data=melted)
plt.title("Distribution des ventes par produit (Boxplot)")
plt.tight_layout()
plt.show()
```
